### Using the Data from Knock Airport Hourly Weather Station Data

https://data.gov.ie/dataset/knock-airport-hourly-weather-station-data

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# we need to skip the first 23 rows so we will skip them and set low memeory to false
df = pd.read_csv("http://cli.met.ie/cli/climate_data/webdata/hly4935.csv", skiprows=23, low_memory=False, nrows=1000)
```

```python
df
```

```python
# convert date column to date time + add new column
df['datetime'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
%matplotlib inline
# remove april because of huge gap
sns.lineplot(x="datetime", y="temp", data=df[1:])
```

## Time series: Simulation

```python
# adapted from https://pandas.pydata.org/pandas-docs/stable/timeseries.html
# 72 hours starting with midnight Jan 1st, 2011

rng = pd.date_range('1/1/2011', periods=72, freq='H')
```

```python
rng
```

```python
ts = pd.DataFrame(np.random.poisson(10, len(rng)), index=rng, columns=['admissions'])
```

```python
ts
```

```python
ts.mean()
```

```python
ts.loc['2011-01-03 16:00:00']
```

```python
ts.loc['2011-01-02 00:00:00':'2011-01-02 23:00:00']
```

```python
ts.loc['2011-01-02 00:00:00':'2011-01-02 23:00:00'].mean()
```

```python
# takes all the variables on the 1 jan and calculates their mean and so on
ts.resample('D').mean()
```
